```python
import os
import numpy as np
import glob
import matplotlib.pyplot as plt
files = sorted(glob.glob('../data/telluric/*.npy'))
files
```

```python
f = files[0]
xx = np.load(f)
x1 = xx[:,0]*1e7
y1 = xx[:,1]

f = files[1]
xx = np.load(f)
x2 = xx[:,0]*1e7
y2 = xx[:,1]

files
```

```python
fig, ax = plt.subplots()
ax.plot(x1,y1,label='Z15PWV05')
ax.plot(x2,y2,label='Z40PWV35')
ax.set_xlabel('Wavelength [A]')
ax.set_ylabel('Flux')
ax.legend()
```
